```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import numpy as np
```

```python
df= pd.read_csv('penguins_size.csv')
df.head()
```

```python
sns.histplot(df['culmen_length_mm'])
```

```python
sns.distplot(df['culmen_depth_mm'])
```

```python
sns.barplot(x='flipper_length_mm	',y='body_mass_g',data=df)
```

```python
df.isnull().any()
```

```python
replace_culmen_length = df['culmen_length_mm'].mean()
replace_culmen_length
```

```python
replace_culmen_depth = df['culmen_depth_mm'].mean()
replace_culmen_depth
```

```python
replace_flipper_length = df['flipper_length_mm'].mean()
replace_flipper_length
```

```python
replace_body_mass = df['body_mass_g'].mean()
replace_body_mass
```

```python
replace_sex = df['sex'].mode()
replace_sex
```

```python
df['culmen_length_mm'].fillna(value=replace_culmen_length, inplace=True)
```

```python
df['culmen_depth_mm'].fillna(value=replace_culmen_depth, inplace=True)
```

```python
df['flipper_length_mm'].fillna(value=replace_flipper_length, inplace=True)
```

```python
df['body_mass_g'].fillna(value=replace_body_mass, inplace=True)
```

```python
df['sex'].fillna(value='MALE', inplace=True)
```
