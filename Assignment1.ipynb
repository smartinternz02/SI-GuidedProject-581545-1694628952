**Import all packages needed**

```python
import pandas as pd
import numpy as np
```

Create all dataframe with 4 columns and 4 rows with features named  `'feature_1', 'feature_2', 'feature_3', 'feature_4'`

```python
features = ['feature_1', 'feature_2', 'feature_3', 'feature_4']
data = np.random.randint(1, 100, size=(4, 4))
df = pd.DataFrame(data, columns=features)

df
```

Rename the column Names to `'Random Value 1', 'Random Value 2', 'Random Value 3', 'Random Value 4'`

```python
df.columns = [f'Random value {i + 1}' for i in range(len(features))]

df
```
